```python
import pandas as pd
import re
from collections import Counter
```

```python
df = pd.read_csv('../make_dictionary_process/stricted_filtered_final_df.csv')
df.tail()
```

```python
journal_ls = df['journal_refer'][df['journal_refer'].notnull()].tolist()
```

```python
testing_ls1 = [i.replace(' : ','') for i in journal_ls]
```

```python
testing_ls2 = [re.sub('[0-9]+ ~ [0-9]+','',i) for i in testing_ls1]
```

```python
journal_keys = [re.sub('\(.+\)','',val) for val in testing_ls2]
```

```python
journal_keys = [','.join(re.findall('[가-힣a-z-A-Z\s]',i)).replace(',','') for i in journal_keys]
```

```python
journal_keys= [i for i in journal_keys if i]
```

```python
for idx,val in enumerate(journal_keys) : 
    try : 
        if val[0] == ' ' : 
            journal_keys[idx] = val[1:]
        if val[-1] == ' ':
            journal_keys[idx] = val[:-1]
    except : 
        print(idx)
```

```python
dict_ = dict(Counter(journal_keys))
```

```python
count_df = pd.DataFrame(index=list(dict_.keys()))
count_df['count'] = list(dict_.values())
```

```python
count_df.sort_values('count',ascending=False).tail(30)
```
