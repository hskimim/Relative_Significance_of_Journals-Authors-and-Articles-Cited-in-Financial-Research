```python
def crawl_the_refer(link_ls):
    citing_and_cited_list_ls = []

    for idx in range(len(link_ls)) : 
        refer_ls = []
        if idx % 200 == 0 : print(idx)

        driver.get(link_ls[idx])
        time.sleep(1)
        req = requests.get(link_ls[idx])
        re = TextResponse(req.url,body=req.text,encoding='utf-8')

        citing_paper_name = re.xpath('//*[@id="printArea"]/div[1]/p/span/text()').extract()[0]
        citing_paper_journal = re.xpath('//*[@id="printArea"]/div[1]/div[1]/div[1]/p/a/text()').extract()[0]\
        .replace('  ','').replace('\r','').replace("\n",'').replace("\t",'')
        citing_paper_year = re.xpath('//*[@id="printArea"]/div[1]/div[1]/p[1]/a/text()').extract()[0]\
        .replace('  ','').replace('\r','').replace("\n",'').replace("\t",'')
        citing_paper_infor = citing_paper_name +' // ' + citing_paper_journal + ' // ' + citing_paper_year

        driver.execute_script("window.scrollTo(100, 10000);")
        cited_times = re.xpath('//*[@id="refeList"]/span/text()').extract()[0].replace('(','').replace(")",'')
        #클릭
        driver.find_element_by_css_selector('#refeList > img').click()
        for i in range(1,100):
            try :
                refer_ls.append(\
        driver.find_element_by_css_selector(\
                                            '#printArea > div.articleBody > div:nth-child(2) > div > p:nth-child({})'\
                                            .format(i)).text)
            except : break
        if refer_ls == [] :
            for i in range(1,100):
                try :
                    refer_ls.append(\
            driver.find_element_by_css_selector(\
                                                '#printArea > div.articleBody > div:nth-child(3) > div > p:nth-child({})'\
                                                .format(i)).text)
                except : break

        citing_and_cited_list_ls.append((citing_paper_infor,refer_ls,cited_times))
        
    return citing_and_cited_list_ls
```

# 나중에 병합할 때 혼류가 없도록 길이에 따른 인덱스를 저장해준다

# 논문의 특성을 나타내는 첫 번째 리스트와 이름이 결합되어 있는 경우가 발생, split 해준다.

\- 기준으로 split

```python
tuned_new_refer = []
for idx in range(len(new_refer_ls)):
    tuned_new_refer.append(new_refer_ls[idx][0].split(" -") + new_refer_ls[idx][1:])
len(tuned_new_refer) == len(new_refer_ls)
```

가장 첫 번째 character 엘리먼트와 3번째 엘리먼스인 년도 사이에 무엇이 있으면 하나로 병합시켜준다. (이름)

```python
merge = [(1, 3)]

for ls in tuned_new_refer : 
    if ls[0] != '0': 
        if not re.findall('[0-9]{1,4}',ls[2]) : 
            for t in merge[::-1]:
                merged = ''.join(ls[t[0]:t[1]])  # merging values within a range
                ls[t[0]:t[1]] = [merged]
```
