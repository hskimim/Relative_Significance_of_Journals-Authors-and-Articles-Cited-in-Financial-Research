```python
paper_name = []
paper_author = []
paper_year = []
paper_name = []
journal_name = []
error_ls = []

for i in double_quote_ls : 
    try : 
        paper_author.append(i[:re.search('“.+”',i).start()])
        paper_year.append(re.findall('20[0-9]{2}|19[0-9]{2}',i))
        paper_name.append(re.findall('“.+”',i))
        journal_name.append(i[re.search('“.+”',i).end():])
    except : 
        error_ls.append(i)
```

```python
double_quote_df = pd.DataFrame()
double_quote_df['author'] =  paper_author
double_quote_df['year'] = paper_year
double_quote_df['paper'] = paper_name
double_quote_df['journal'] = journal_name
double_quote_df
```

```python
double_quote_df['journal'] = \
[re.sub('\([0-9]{4}\)|[0-9]{4}|\([0-9]{1,2}\)|[.]| [0-9]{1,2}','',i).strip() for i in double_quote_df['journal'].tolist()]
double_quote_df.tail()
```

```python
def count_under_col(val,list_=False,how_much=10,df=False):
    if list_ : 
        dict_ = Counter([i[0].lower() for i in double_quote_df[val].tolist() if i])
    else : dict_ = Counter([i for i in double_quote_df[val].tolist() if i])
    count_df = pd.DataFrame.from_dict(dict_,orient='index',columns=['count'])
    if df : 
        return count_df.sort_values('count',ascending=False)
    return count_df.sort_values('count',ascending=False).head(how_much)
```

```python
display(count_under_col('paper',list_=True,how_much=10))
display(count_under_col('author',list_=False,how_much=10))
display(count_under_col('journal',list_=False,how_much=10))
display(count_under_col('year',list_=True,how_much=10))
```
