2002,2003년도에는 파일이 hwp였고, 이에 따라서, 기존에 rule-based로 지정되었던, pdf 특유의 double quote가 적용되지 않고 있었다. 이에 대한 함수를 따로 만들어주는 것보다는, quotation 을 pdf방식으로 바꿔준 후, 기존의 함수에 적용하는 것이 더욱 효율적이라고 판단, 이를 바꿔준 후, fm_sent_ls를 재정의(re-define)하는 과정을 진행한다.

```python
fix_fm_ls = []
revised_fm_sent_ls = []
for idx,val in enumerate(fm_sent_ls) : 
    if "한국재무관리학회/2002" in val[1] or "한국재무관리학회/2003" in val[1] : 
        fix_fm_ls.append(val)
    else : revised_fm_sent_ls.append(val) #2002,2003 년을 제외한 fm_sent_ls,revised_fm_sent_ls을 선언합니다.
        
fm_2002_2003_p_ls = [] # ".+"을 가진 참고문헌이 2002,2003년에서는 quote_ls입니다.
fm_2002_2003_ip_ls = []

for idx,i in enumerate(fix_fm_ls) : 
    if re.search('".+"',i[0]) :
        fm_2002_2003_p_ls.append(i)
    else : fm_2002_2003_ip_ls.append(i) # fm_2002_2003_ip_ls는 hwp 방식의 quotation이 없는 참고문헌을 모아 둔 list이다.

process_ls = [i[0] for i in fm_2002_2003_p_ls]
m_ls= [[m.span() for m in re.finditer('".+"',i)] for i in process_ls]

first_quote_ls = [i[0][0] for i in m_ls]
second_quote_ls = [i[0][1] for i in m_ls]

final_ls = []

for idx,val in enumerate(process_ls) : 
    final_ls.append(\
val[:first_quote_ls[idx]] + '“' + val[first_quote_ls[idx]:second_quote_ls[idx]] + '”' + val[second_quote_ls[idx]:])
    
revised_fm_sent_ls += list(zip([i.replace('"','') for i in final_ls],[i[1] for i in fm_2002_2003_p_ls])) 
revised_fm_sent_ls += fm_2002_2003_ip_ls # 후에 make_imperfect_ls에서 ip_ls를 따로 선언해줄 것이기 때문에, 더해도 무방하다.
assert len(fm_sent_ls) == len(revised_fm_sent_ls)
fm_sent_ls = revised_fm_sent_ls.copy() #fm_sent_ls의 재정의 프로세스
```

에러파일은 자체 path 내에 년도가 없다. 이에 따라서, 역추적해서 year을 찾아, 딕셔너리로 할당한 후, 리바이싱한다.

```python
appending_dict = {}
for file in os.listdir("../dictionary_methodology/error_solved_file/") : 
    for path in fm_path + f_path + sc_path + dr_path : 
        if file in path : 
             appending_dict["error_solved_file/"+file] = path.split("/")[-2]
                
for idx in range(len(dictionary_df['citing_year'])) : 
    if not re.search('20[0-9]{2}',dictionary_df['citing_year'][idx]) : 
        dictionary_df['citing_year'][idx] = appending_dict[dictionary_df['citing_year'][idx]]
```

```python
def catch_the_citing_paper_name(new_dictionary_df,citing_df) : 
    total_count = 0
    citing_paper_name = []
    citing_paper_author = []
    title_dict = {}
    for path_idx,path in enumerate(new_dictionary_df['path']) :
        if path_idx % 4000 == 0 : print(path_idx)
        with open(path) as f : 
            lines = f.readlines()
            lines = [i.replace("\n",'').replace('\t','').replace('\ufeff','').replace("·",',').replace('*','') for i in lines[:50] if re.search("[a-zA-Z가-힣]",i)]
            new_lines = ','.join(lines[:4]).replace(",",' ')
            lines = ','.join(lines[:4]).replace(",",' ').replace(' ','')
        for idx in list(citing_df[citing_df['from'] == dictionary_df['from'][path_idx]].index) : 
            count = 0
            if citing_df['paper'][idx].replace(" ",'') in lines : 
                citing_paper_name.append(citing_df['paper'][idx])
                citing_paper_author.append(citing_df['author'][idx])
                count += 1
                total_count += 1
                break
        if count == 0 : 
            citing_paper_name.append('None')
            citing_paper_author.append('None')
    
    new_dictionary_df['citing_paper_name'] = citing_paper_name
    new_dictionary_df['citing_paper_author'] = citing_paper_author
    
    print('전체 데이터의 갯수 {0} 에서 {1} 개를 포착했습니다, {2} %.'.format(len(new_dictionary_df),total_count,\
                                                   np.round(total_count*100/len(new_dictionary_df))))
    return new_dictionary_df

citing_df = pd.read_csv("citing_paper.csv")
%time new_dictionary_df = catch_the_citing_paper_name(dictionary_df,citing_df)
if  len(new_dictionary_df) == len(dictionary_df) : print('perfect')
```

```python
import concurrent.futures

ip_ls = [[i[0] for i in fm_ip_ls],
          [i[0] for i in f_ip_ls],
          [i[0] for i in sc_ip_ls],
          [i[0] for i in dr_ip_ls]]

cat_ls = list(zip([quote_df for i in range(len(ip_ls))],ip_ls))
len(cat_ls)
```

```python
def main() : 
    final_df = []
    with concurrent.futures.ProcessPoolExecutor(max_workers = 8) as executor:
        for _,df_ in zip(cat_ls,executor.map(pars.catch_the_paper_under_similarity_score_for_concurrent, cat_ls)):
            final_df.append(df_)
    return final_df

if __name__ == '__main__':
    %time final_df = main()
```
