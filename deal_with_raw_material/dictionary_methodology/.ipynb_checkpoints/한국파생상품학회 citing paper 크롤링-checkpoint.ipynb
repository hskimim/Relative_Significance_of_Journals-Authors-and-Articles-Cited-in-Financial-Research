```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
year_ls = [str(i) for i in range(2002,2019)]
book_num_ls = [i for i in range(1,10)]
```

```python
header = {
    'User-Agent': 
    'Mozilla/5.0 (X11; Linux x86_64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/73.0.3683.86 Safari/537.36'
}
cookie = {
    'Cookie': 'clientName=%ED%95%9C%EA%B5%AD%ED%8C%8C%EC%83%9D%EC%83%81%ED%92%88%ED%95%99%ED%9A%8C%28%EA%B5%AC+%ED%95%9C%EA%B5%AD%EC%84%A0%EB%AC%BC%ED%95%99%ED%9A%8C%29; PdfOK=True; clientKey=3022; ASPSESSIONIDQQSDDBAT=FHJBBIMAPKDHGFAGHHKPAFDL; ASPSESSIONIDQQQAADBS=ANIJPDJBNEEBADCHJFPOAMPH'
}
process_df = pd.DataFrame()

for year in year_ls : 
    for book_num in book_num_ls : 
        datas = {
            'strInst': '3022',
            'strPubl': '648',
            'cate1': '{}'.format(year),
            'cate2': '{},{}'.format(int(year[2:]) + 8,book_num),
            'stryear': '{}'.format('2006'),
            'strvol': '{}'.format('14'),
            'strnum': '{}'.format('2'),
        }


        req = requests.post('http://journal.kstudy.com/ISS_RForm.asp',headers=header,data=datas,cookies=cookie)
        req.encoding = 'utf-8'
        dom = BeautifulSoup(req.text,'lxml')

        paper_ls = [i.text for i in dom.select('#contents > div > div > div.title_info > ul > li:nth-of-type(1) > a')]
        author_ls = [i.text for i in dom.select('#contents > div > div > div.title_info > ul > li:nth-of-type(2)')]

        if len(paper_ls) == len(author_ls) :
            if len(paper_ls) !=0 :
                zip_dict = dict(zip(paper_ls,author_ls))
                print(len(zip_dict))
            else : print('error')
    
    process_df = process_df.append(pd.DataFrame.from_dict(zip_dict,orient = 'index'))
```

```python
process_df.to_csv('파생상품학회 citing paper.csv')
```
